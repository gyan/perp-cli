```python
import pandas

total_sum = 0 

columns = ["name","timestamp","amm","trader","exchangedPositionSize","side","spotPrice","fundingPayment","margin","positionNotional","fee","positionSizeAfter","realizedPnl","unrealizedPnlAfter","badDebt","liquidationPenalty","address","txHash","blockNumber"]
df = pandas.read_csv('eth_data.csv', names=columns)
df= df.sort_values("timestamp", axis=0, ascending=False)

```

```python
df2 = df.copy(deep=True)
```

```python
from web3 import Web3
# w3 = Web3(Web3.HTTPProvider('https://rpc.xdaichain.com/'))
w3 = Web3(Web3.WebsocketProvider('wss://rpc.xdaichain.com/wss'))

for ix, item in df2.iterrows():
    if (str(item['epoch']) == 'nan'):
        df2.at[ix, 'epoch'] = w3.eth.getBlock(item['blockNumber']).timestamp
        print(ix)
    else:
        print("already updated at index: ", ix)
# df2.head()
# w3.eth.getTransaction("0x8128d0bd3d7f41f92a9da39fd324ec0febbee4013bb3172587f8651bfb0907f0") 
```
